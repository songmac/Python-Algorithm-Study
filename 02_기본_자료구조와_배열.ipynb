## 리스트와 튜플의 개념

- 리스트 : 원소 변경 가능 (mutable)
- 튜플 : 원소 변경 불가능 (immutable)

- mutable vs immutable? 
    - mutable 자료형 : 리스트, 딕셔너리, 집합 등이 있으며 값을 변경할 수 있음
    - immubatle 자료형 : 수, <b>문자열</b>, 튜플 등이 있으며 값을 변경할 수 없음

### 슬라이스식으로 원소에 접근하기

s[i:j] : s[i] 부터 s[j-1] 까지 나열

s[i:j:k] : s[i] 부터 s[j-1] 까지 k씩 건너뛰며 나열

- i, j > len(s) 이면 len(s) 가 지정된 것으로 간주
- i가 없거나 None 이면 0으로 지정된 것으로 간주
- j가 없거나 None 이면 len(s)가 지정된 것으로 간주

```python
# 슬라이스식으로 원소 꺼내기

s = [11, 22, 33, 44, 55, 66, 77]
print(s[0:7])
print(s[0:7:2])
print(s[3:1])  # 슬라이스 범위가 잘못되었지만 오류가 나지는 않음
print(s[3:-2])  
print(s[-1:-3]) # 슬라이스 범위와 방향이 잘못되었지만 오류가 나지는 않음
print(s[-1:-3:-1])  # 슬라이스 방향을 바꿔주면 정상적으로 슬라이싱이 됨
```

## 자료구조의 개념

- 자료구조 : 데이터 단위와 데이터 자체 사이의 물리적 또는 논리적인 관계
- 자료구조를 알아야 하는 이유 : 컴퓨터에서 처리해야 하는 많은 데이터를 모아 효율적으로 관리하고 구조화해야 함

```python
# 비교 연산자로 배열의 대소 또는 등가 관계 판단하기

print([1, 2, 3] == [1, 2, 3])  # 맨 앞 원소부터 차례로 비교
print([1, 2, 3] < [1, 2, 4])  # 어떤 원소의 값이 크면 그 배열이 큰 것으로 판단
print([1, 2, 3, 4] <= [1, 2, 3, 4])
print([1, 2, 3] < [1, 2, 3, 5])  # 배열의 원소 수가 다른 경우에는 원소 수가 많은 배열을 더 크다고 판단
print([1, 2, 3] < [1, 2, 3, 5] < [1, 2, 3, 5, 6])
```

> 참고. 등가성과 동일성

- 등가성 비교는 == 을, 동일성 비교는 is 를 사용
    - 등가성(equlity)? 값이 같은가를 비교
    - 동일성(identity)? 같은 객체인가를 비교. 즉, 메모리 주소가 같은지를 비교

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)  # a와 b는 같은 원소를 가지고 있으므로 True
print(a is b)  # a와 b는 다른 객체이므로 False
```

```python
# 파이썬에서의 등가성, 동일성 비교
x = 1000
y = 1000
print(x == y)  # x와 y는 같은 값을 가지므로 True
print(x is y)  # x와 y는 다른 객체이므로 False

# 파이썬에서의 예외 (예시)
a = 10
b = 10
print(a == b)  # a와 b는 같은 값을 가지므로 True
print(a is b)  # 파이썬은 작은 숫자(-5~256)나 짧은 문자열 등을 캐싱해서 동일한 객체를 재사용 하기도 함 # True
```

## 배열 원소의 최댓값 구하기

```python
# 최댓값을 구하는 과정

a = [22, 57, 11, 91, 32]

def max_of(a):
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:  # a[1] ~ a[4] 까지 배열 원소를 하나씩 차례대로 주목하여 살펴보는 방식 : 스캔 (scan) / 트래버스(traverse)
            maximum = a[i]
    return maximum

max_of(a)
```

## 배열 원소의 최댓값을 구하는 함수 구현하기

image.png

```python
 # [Do it! 실습 2-2] 시퀀스 원소의 최댓값 출력하기

from typing import Any, Sequence  # typing 모듈을 쓰면 다양한 입력 타입을 하나로 처리 할 수 있게 함 (타입 힌트로 협업에 도움)
                                  # 실제로는 list, int, str 등 내장 타입을 직접 작성해도 됨

def max_of(a: Sequence) -> Any:  # 함수 어노테이션(Function Annotation) : 파이썬 함수의 매개변수나 반환값에 대해 타입 정보를 명시하는 문법 (강제성 X, 코드 자체에 영향도 X)
                                 # def 함수이름(매개변수: 타입) -> 반환타입:
                                 #  - Any : 제약이 없는 임의의 자료형을 의미
                                 # - Sequence : 시퀀스 형을 의미 ex. 리스트(list)형, 바이트 배열(bytearray)형, 문자열(str)형, 튜플(tuple)형, 바이트열(bytes)형

    """시퀀스형(Sequence) a 요소의 최댓값을 임의의 자료형(Any)으로 반환""" 
    maximum = a[0]
    for i in range(1, len(a)):
         if a[i] > maximum:
            maximum = a[i]
    return maximum

# 값을 입력받아 max_of()함수를 실행
# 스크립트 프로그램이 직접 실행될 때 변수 __name__은 '__main__'임
# 스크립트 프로그램이 임포트 될 때 변수 __name__은 원래의 모듈 이름임
if __name__ == '__main__':
    print('배열의 최댓값을 구합니다.')
    num = int(input('원소 수를 입력하세요 : '))
    x = [None] * num    # 원소 수가 num인 리스트를 생성

    for i in range(num):
        x[i] = int(input(f'x[{i}]를 입력하세요.: '))

    print(f'최댓값은 {max_of(x)}입니다.')
```

## 재사용할 수 있는 모듈 작성하기

- 하나의 스크립트 프로그램을 모듈(module)이라고 함
- 확장자(.py)를 포함하지 않는 파일의 이름 자체를 모듈 이름으로 사용
- 연산자 __name__ : 모듈 이름을 나타내는 변수
- 모듈 객체에는 name 외에도 loader, package, spec, path, file 등이 있음

```python
%%writefile max.py

from typing import Any, Sequence

def max_of(a: Sequence) -> Any:
    """시퀀스형(Sequence) a 요소의 최댓값을 임의의 자료형(Any)으로 반환"""
    maximum = a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximum = a[i]
    return maximum
```

```python
# [Do it! 실습 2-3] 배열 원소의 최댓값을 구해서 출력하기 (원솟값을 입력받음)

from max import max_of # max.py을 실행하려는 스크립트 파일과 같은 경로에 저장해야 함

print('배열의 최댓값을 구합니다.')
print('주의: "End"를 입력하면 종료합니다.')

number = 0
x = [] # 빈 리스트

while True:
    s = input(f'x[{number}]를 입력하세요.: ')
    if s == 'End': # max.py 내에서는 실행되지 않음('__name__==__main__'이 성립하지 않기 때문)
        break # while문 종료
    x.append(int(s)) # 배열의 끝에 추가
    number += 1 # 정수를 입력 받을 때마다 1씩 증가

print(f'{number}개를 입력했습니다.')
print(f'최댓값은 {max_of(x)}입니다.')
```

## 배열 원소를 역순으로 정렬하기

- for i in range(n // 2):
- a[i]와 a[n -i -1]의 값을 교환 (i = 0, 1, 2, ...)

```python
# [Do it! 실습 2-6] 뮤터블 시퀀스 원소를 역순으로 정렬

from typing import Any, MutableSequence

def reverse_array(a: MutableSequence) -> None:
    """뮤터블 시퀀스형 a의 원소를 역순으로 정렬"""
    n = len(a)
    for i in range(n // 2):
         a[i], a[n -i -1] = a[n -i -1], a[i]

if __name__ == '__main__':
    print('배열 원소를 역순으로 정렬합니다.')
    nx = int(input('원소 수를 입력하세요.: '))
    x = [None] * nx   # 원소 수가 nx인 리스트를 생성

    for i in range(nx):
        x[i] = int(input(f'x[{i}]값을 입력하세요 : '))

    reverse_array(x)  # x를 역순으로 정렬

    print('배열 원소를 역순으로 정렬했습니다.')
    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
```

## 10진수, 8진수, 16진수 의 기수 살펴보기
- n진수는 n을 기수로 하는 수
- 10진수 : 0 ~ 10 까지의 숫자 10종류 사용
- 8진수 : 0 ~ 7 까지의 숫자 8종류 사용
- 16진수 : 0 ~ 9, A ~ F(a~f) 까지의 숫자 & 알파벳 16종류 사용

```python
# Do it! 실습 2-7 [A] 10진수 정수값을 입력받아 2~36진수로 변환하여 출력하기

# 함수 정의
def card_conv(x: int, r: int) -> str:
    """정수 x 를 r 진수로 변환한 뒤 그 수를 나타내는 문자열을 반환"""

    d = ''  # 변환 뒤 문자열
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ' # 36진수를 표현

    while x > 0:  # 진법 변환에서는 몫이 0이 될 때까지 계속 나머지를 구하면서 몫을 갱신하는 과정을 반복 
        d += dchar[x % r]  # 'x를 r로 나눈 후 나머지'에 해당하는 문자를 꺼내 결합 
        x //= r # 'x를 r로 나눈 후 몫'을 x에 집어 넣음

    return d[::-1] # 역순으로 반환

# 실행 예시
print(card_conv(59, 16))  # 16으로 나눈 나머지 11(B), 나머지 3(3)에 해당하는 문자열로 각각 변환되었고, 역순으로 출력됨
```

```python
# 10진수 정수값을 입력받아 2~36진수로 변환하여 출력하기 (실습 2-7[A] 수정)
# 진법 변환 과정을 print() 함수로 출력하는 방식으로 변경

def card_conv(x: int, r: int) -> str:
    """정수 x를 r 진수로 변환한 뒤 그 수를 나타내는 문자열을 반환"""

    d =  ''  # 변환 뒤 문자열
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    n = len(str(x))  # 변환하기 전의 자릿수

    print(f'{r:2} | {x:{n}d}')
    while x > 0: 
        print('   +' + (n + 2) * '-')
        if x // r:
            print(f'{r:2} | {x // r:{n}d} … {x % r}')
        else:
            print(f'     {x // r:{n}d} … {x % r}')
        d += dchar [x % r]  # 해당하는 문자를 꺼내 결합
        x //= r

    return d[::-1]  # 역순으로 반환

# 이하 Do it! 실습 2-7 [B]와 동일

if __name__ == '__main__':
    print('10진수를 n진수로 변환합니다.')

    while True:
        while True :  # 음이 아닌 정수를 입력받음
            no = int(input('변환할 값으로 음이 아닌 정수를 입력하세요.: '))
            if no > 0:
                break

        while True :  # 2~36진수의 정수값을 입력받음
            cd = int(input('어떤 진수로 변환할까요?: '))
            if 2 <=  cd <=  36:
                break

        print(f'{cd}진수로는 {card_conv(no, cd)}입니다.')

        retry = input( "한 번 더 변환할까요?(Y … 예/N … 아니오) : ")
        if retry in {'N', 'n'}:
           break
```

        즉, c는 변수 = 값 그 자체 (메모리에 직접 저장), 포인터 써야만 참조 공유 가능 <br>
        python은 변수 = 객체에 대한 참조, 자주 쓰이는 작은 정수만 미리 만들어둔 객체를 재사용.

```python
# [Do it! 실습 2C-5] 1부터 n까지 정수의 합 구하기 (이뮤터블인 경우)
# 1부터 n까지 정수의 합을 구해 반환하는 sum1ton() 함수를 구현

def sum_1ton(n):
    """1부터 n까지 정수의 합"""
    sum = 0
    while n > 0:
        sum += n
        print(f'현재 n = {n}, sum = {sum}')  # 현재 n과 sum의 값을 출력
        print(f'현재 n의 id = {id(n)}, sum의 id = {id(sum)}')  # 현재 n과 sum의 id를 출력하여 n이 어떤 객체를 가리키는지 확인
        n -= 1
    return sum

x = int(input('x의 값을 입력하세요.: ')) # 변수명을 함수와 다르게 지정하는 이유? 입력값과 로컬 변수의 역할을 구분!
print(f'x의 id ? {id(x)}')  # x의 id를 출력하여 x가 어떤 객체를 가리키는지 확인
print(f'1부터 {x}까지 합은 {sum_1ton(x)}입니다.')

# 출력 결과 : 이뮤터블 객체인 int형의 경우, 함수 내 인수로 전달된 n의 값이 변경되더라도 x의 값은 변경되지 않음 
```

## 소수 나열하기
- 정수 이하의 소수(prime number)을 모두 나열하는 알고리즘
- 소수 : 자신과 1 이외의 정수로 나누어 떨어지지 않는 정수
- 정수 n이 소수일 조건 : 2부터 n-1까지 어떤 정수로도 나누어 떨어지지 않음
- 만약 나누어 떨어지는 정수가 하나 이상 존재하면 그 수는 합성수(composite number)임
- 소수를 구하는 부분은 이중 for문 구조로 작성

```python
# [Do it! 실습 2-8] 1,000 이하의 소수를 나열하기

counter = 0 # 나눗셈 횟수를 카운트
# prime = []  # 소수를 저장할 리스트

for n in range(2, 1001):  # 2 ~ 1000까지 순회
    for i in range(2, n):  # 2 ~ n-1 까지 나누어 떨어지는 수가 있는지 확인
        counter += 1
        if n % i == 0:  # 나누어 떨어지면 소수가 아님
            break
    else:
        print(n, end =', ')  # 소수인 경우 출력
        # prime.append(n)

print(f'\n나눗셈을 실행한 횟수: {counter}')

print(f'----'* 10)

# print(f'소수 리스트 : {prime}')
# print(f'총 소수의 개수 : {len(prime)}개')
```

> 참고. 알고리즘 개선하기 1 : 나눗셈 횟수 줄이기

- n이 2와 3으로 나누어 떨어지지 않는다면 2X2인 4와 2X3인 6으로도 나누어 떨어지지 않음(불필요한 나눗셈을 계속 실행)
- 따라서, 7이 소수인지는 7보다 작은 소수인 2,3,5로 나눗셈을 실행하는 것만으로도 충분
- 소수를 구하는 과정에서 지금까지 구한 소수를 배열 prime의 원소로 저장하도록 개선
- n이 소수인지 판단할 때 배열 prime에 저장한 소수로 나눗셈을 하면 됨
- 2는 소수인 것이 분명하므로 2를 배열의 첫 원소인 prime[0]에 저장

```python
# [Do it! 실습 2-9] 1,000 이하의 소수를 나열하기(알고리즘 개선 1)

counter = 0           # 나눗셈 횟수
ptr = 0               # 이미 찾은 소수의 개수 (초기값 :0)
prime = [None] * 500  # 소수를 저장하는 배열
                      # 500개인 이유? 짝수는 소수가 아니므로 전체 개수(1,000개)의 반으로 모든 소수를 배열에 넣을 수 있기 때문
# n = 2
prime[ptr] = 2        # 2는 소수이므로 초깃값으로 지정
ptr += 1              # 이미 찾은 소수의 개수 : 1개

# n = 3 ~ 1000 (중에서 홀수만)
for n in range(3, 1001, 2):  # 3부터 1000까지 홀수만 순회
    for i in range(1, ptr):  # 이미 찾은 소수로 나눔 (1 부터 ptr-1 까지 반복)
                             # i가 1부터 시작하는 이유? 판단 대상인 n이 홀수이므로, 0으로 나누는 의미가 X
        counter += 1
        if n % prime[i] == 0:  # 소수 배열 prime 을 순회하면서, 모든 소수로 나누는 시도
            break        
    else:  # 끝까지 나누어 떨어지지 않았다면 소수로 배열에 등록
        prime[ptr] = n  # -> 3(counter:0), 5(counter:1 by 3), 7(counter:2 by 3, 5) ...etc. 
        ptr += 1  # 소수 배열 추가 
                      
# ptr의 소수를 출력
for i in range(ptr):  
    print(prime[i], end=", ")
print()
print(f'나눗셈을 실행한 횟수: {counter}')  # 나눗셈을 실행하는 횟수가 78,022 -> 14,622로 크게 줄음
```

>참고. 알고리즘 개선하기 2 : 직사각형 대칭구조 아이디어 적용하기
- 넓이가 100인 직사각형의 가로, 세로 변의 길이가 있다면, 모든 직사각형은 정사각형 10X10을 경계로 대칭 구조를 이룸
- 직사각형 한 변의 길이만 나눗셈을 시도하고, 그 과정에서 한번도 나누어 떨어지지 않으면 소수라고 판단해도 좋음
- 즉 어떤 정수 n은 'n의 제곱근 이하의 어떤 소수로도 나누어 떨어지지 않음'

```python
# [Do it! 실습 2-10] 1,000 이하의 소수를 나열하기(알고리즘 개선 2)

counter = 0  # 곱셈과 나눗셈을 합한 횟수
ptr = 0  # 이미 찾은 소수의 개수
prime = [None] * 500  # 미리 빈 배열을 선언해줘야 하는 이유? prime 배열에 n을 자동할당하는 게 아니라 직접 할당하는 방식 이기 때문 ! 
# prime = []  # 이렇게 표현할 경우 prime[0] = 2 했을 때 IndexError 발생 ! (append로 해결 가능하긴 함) 

prime[ptr] = 2  # 2는 소수
ptr += 1

prime[ptr] = 3  # 3은 소수
ptr += 1

for n in range(5, 1001, 2):  # 홀수만을 대상으로 설정
    i = 1
    while prime[i] * prime[i] <=  n:
        counter += 2  # 두 연산(곱셈, 나눗셈)의 실행 횟수를 셈
        if n % prime[i] == 0:  # 나누어 떨어지므로 소수가 아님
            break  # while문 중단
        i += 1
    else:  # while문이 정상 종료되면 실행. (단, break 로 중간에 빠져나가면 else문은 무시됨!)
        prime[ptr] = n  # 끝까지 나누어 떨어지지 않았다면 소수로 배열에 등록
        ptr += 1
        counter += 1  # 여기에 count를 하는 이유? 
                      # 소수 판별 과정에서의 마지막 나눗셈(if n % prime[i] == 0)을 포함
                      # 1) while 루프 안 : prime[i] * prime[i] 검사 → 나눗셈
                      # 2) while 루프 종료 : prime[i] * prime[i] > n 되면 False로 바로 나옴, n % prime[i] 안 함
                      # 3) else : 마지막 나눗셈 안 했으므로, counter 만 수동으로 +1

for i in range(ptr):  # ptr개의 소수를 출력
    print(prime[i], end=', ')

print()
print(f'곱셈과 나눗셈을 실행한 횟수: {counter}')   # 나눗셈을 실행하는 횟수가 14,622-> 3,774로 크게 줄음
```

```python
# [Do it! 실습 2-10] 1,000 이하의 소수를 나열하기(알고리즘 개선 2) - 배열의 원솟수를 미리 결정하지 않음

counter = 0             # 곱셈과 나눗셈을 합한 횟수
prime = [2, 3]          # 소수를 저장하는 배열

for n in range(5, 1001, 2):     # 홀수만을 대상으로 설정
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:   # 나누어 떨어지므로 소수가 아님
            break               # 반복 중단
        i += 1
    else:                       # 끝까지 나누어 떨어지지 않았다면
        prime += [n]            # 소수로 배열에 등록
        counter += 1

for i in prime:                 # 소수를 출력
    print(i)
print(f'곱셈과 나눗셈을 실행한 횟수: {counter}')
```

## 리스트의 원소와 복사

- 파이썬에서 변수는 객체와 연결된 이름에 불과하기 때문에 리스트, 튜플의 원소 자료형을 미리 정할 필요가 없음

```python
# [Do it! 실습 2C-7] 자료형을 정하지 않은 리스트 원소 확인하기

x = [15, 64, 7, 3.14, [32, 55], 'ABC']
for i in range(len(x)):
    print(f'x[{i}] = {x[i]}')  # x[i] : 각각 int형, int형, int형, float형, list형, str형의 객체를 참조하는 이름
```

- 리스트의 얕은 복사(shallow copy) : 리스트를 복사할 때 사용하는 copy()함수는 주의해서 사용해야 함. 리스트를 복사한 후 원솟값을 변경하면 복사된 원솟값까지 변경될 수 있음
- 즉, 참조 값만 복사함

```python
x = [[1, 2, 3], [4, 5, 6]]
y = x.copy()
x[0][1] = 9
x
```

```python
y
```

- 리스트의 깊은 복사(deep copy)/전체복사 : 위 같은 상황을 피하려면 구성 원소 수준으로 복사하는 방법이 필요
- 즉, 참조 값 뿐만 아니라 참조하는 객체 자체를 복사함

```python
import copy
x = [[1, 2, 3], [4, 5, 6]]
y = copy.deepcopy(x)
x[0][1] = 9
x
```
