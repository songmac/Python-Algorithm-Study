```python
# 책에 있는 코드 이해하기 (solving)

def med3(a, b, c):
  if a >= b:
    if b >= c:
      return b
    elif a <= c:
      return a
    else:
      return c
  elif a > c:
    return a
  elif b > c:
    return c
  else:
    return b

print('세 정수이 중앙값을 구합니다.')

a = int(input('정수 a의 값을 입력하세요.: '))
b = int(input('정수 b의 값을 입력하세요.: '))
c = int(input('정수 c의 값을 입력하세요.: '))

print(f'중앙값은 {med3(a,b,c)}입니다.')
```

## 실습 1-7. 1부터 n까지 정수의 합 구하기 (while문)
- while문 반복 : 어떤 조건이 성립하는 동안 반복해서 처리하는 것을 반복 구조(repetition structure)라 하고 일반적으로 루프(loop)라고 함

image.png



```python
print('1부터 n까지 정수의 합을 구합니다.')
n = int(input('n값을 입력하세요.: '))

sum = 0
i = 1

while i <= n:  # i가 n보다 작거나 같은 동안 반복
    sum += i   # sum에 i를 더함
    i += 1     # i에 1을 더함

print(f'1부터 {n}까지 정수의 합은 {sum}입니다.')
print(f'i값은 {i}입니다.') #n이 5일 때, 반복문을 마친 뒤 i 값은 6
```

## 실습 1-8. 1부터 n까지의 합 구하기 2 (for 문)


- 변수가 하나만 있을 때는 while문 보다 for문을 사용하는 것이 좋음

```python
print('1부터 n까지의 합을 구합니다.')
n = int(input('n값을 입력하세요.: '))

sum = 0
for i in range(1, n + 1):
    sum += i  # sum에 i를 더함
    #sum = n * (n+1) // 2

print(f'1부터 {n}까지 정수의 합은 {sum}입니다.')
```

## 실습 1-9. a부터 b까지 정수의 합 구하기(for 문)
- 연속하는 정수의 합을 구할 때 시작하는 값이 1 이 아닌 정수를 입력 받았다면 range() 함수에 전달할 시작값과 끝값을 오름차순으로 정렬해야 함
- a와 b를 교환할 때 사용한 단일 대입문은 튜플임
image.png

```python
print('a부터 b까지의 합을 구합니다.')
a = int(input('정수 a를 입력하세요.: '))
b = int(input('정수 b를 입력하세요.: '))

if a > b:
    a, b = b, a  # a와 b를 오름차순으로 정렬

sum = 0
for i in range(a, b + 1):
    sum += i  # sum에 i를 더함

print(f'{a}부터 {b}까지 정수의 합은 {sum}입니다.')
```

## 실습 1-10. a부터 b까지 정수의 합 구하기 1


```python
print('a부터 b까지의 합을 구합니다.')
a = int(input('정수 a를 입력하세요.: '))
b = int(input('정수 b를 입력하세요.: '))

if a > b:
    a, b = b, a

sum = 0
for i in range(a, b + 1):  # b - a번 반복
    if i < b:              # i가 b보다 작으면 합을 구하는 과정을 출력
                           # 더하는 수가 n개 이면 + 기호는 n - 1개 출력해야 함
        print(f'{i} + ', end='')
    else:                  # i가 b보다 크거나 같으면 최종값 출력을 위해 i =를 출력
        print(f'{i} = ', end='')
    sum += i               # sum에 i를 더함

print(sum)
```

## 실습 1-12. +와 -를 번갈아 출력하기 1 (★★★)

```python
# 산술 연산자
# / : 나누기를 의미 (소수점까지 표시)
# // : 몫을 의미 (소수점 뒤를 한 번 더 잘라낸다고 생각하기)
# % : 나머지를 의미


print('+와 -를 번갈아 출력합니다.')
n = int(input('몇 개를 출력할까요?: '))

for i in range(n):          # 반복 n번 (n : 0 ~ 10)
    if i % 2:
        print('-', end='')  # i가 홀수인 경우 - 출력 # 나머지 값이 있을 경우 True!
    else:
        print('+', end='')  # i가 짝수인 경우 + 출력

print()
```

## 실습 1-14. 반복 과정에서 조건 판단하기 3



```python
# *를 n개 출력하되 w개마다 줄바꿈하기1

print('*를 출력합니다.')
n = int(input('몇 개를 출력할까요?: '))
w = int(input('몇 개마다 줄바꿈할까요?: '))

for i in range(n):
  print('*', end='') #*을 출력 후 더이상 나누어 지지 않을 때, 조건 2 수행
  if i % w == w - 1: #조건2 : (n/w 한 나머지) == (w-1)
                     #컴퓨터는 숫자를 0부터 셈
    print()

if n % w == 0: #'if n % w :' 와 표현 동일(조건 True)
               #for 문 수행과 동시에 조건 판단
               # 조건1 : *를 출력 하되, 나누어지는 수일 때 띄어쓰기 실행

  print()

# <결과 리뷰>
#(14, 5) : '*' 5개 2줄, 4개 1줄
#(31, 11) :  '*' 11개 2줄, 9개 1줄
#(56, 15) : '*' 15개 3줄, 11개 1줄
#(79, 20) : '*' 20개 3줄, 19개 1줄
#(21, 1) : '*' 21개 1줄 , 0개 0줄
#(3, 10) : '*' 3개 0줄
#(0, 10) : X (0으로 수렴)
#(10, 0) : error (무한대로 발산)
```

```python
#코드 변형

print('*를 출력합니다.')
n = int(input('몇 개를 출력할까요?: '))
w = int(input('몇 개마다 줄바꿈할까요?: '))

for i in range(n):
  print('*', end='')

  if n % w == 0:
    print()
  elif i % w == w - 1:
    print()
```

# 실습 1-15. *를 n개 출력하되 w개마다 줄바꿈하기 2

```python
print('*를 출력합니다.')
n = int(input('몇 개를 출력할까요?: '))
w = int(input('몇 개마다 줄바꿈할까요?: '))

for _ in range(n // w):  # 반복 n // w번 반복 #변수를 쓰지 않아 '_'로 자리를 대체함
    print('*' * w)

rest = n % w
if rest:
    print('*' * rest)  # if 문 1번 판단
```

# 실습 1-16. 1부터 n까지 정수의 합 구하기(n값은 양수만 입력받음)

- 실습 1-8 프로그램에서 n이 음수 일 때 적용되지 않으므로 이를 개선

```python
print('1부터 n까지 정수의 합을 구합니다.')

while True:
    n = int(input('n값을 입력하세요.: '))
    if n > 0:
        break  # n이 0보다 커질 때까지 반복
               # 파이썬은 사후 판단 반복문(do-while, repeat~until) 등을 제공하지 않음

sum = 0
i = 1

for i in range(1, n + 1):
    sum += i  # sum에 i를 더함
    i += 1    # i에 1을 더함

print(f'1부터 {n}까지 정수의 합은 {sum}입니다.')
```

- for 문에서 반복 종료 시 i는 n

# 실습 1-17. 가로 세로가 정수이고 넓이가 area인 직사각형에서 변의 길이를 나열하기
image.png


```python
area = int(input('직사각형의 넓이를 입력하세요.: '))

for i in range(1, area + 1):  # 1부터 사각형의 넓이 계산
    if i * i > area: break  # key point! break 문이 실행되지 않았을 때만 else문이 실행됨
    if area % i: continue  # area 약수를 나열
    print(f'{i} × {area // i}')  #짧은 변, 긴 변 순서대로 출력
```

```python
print('오른쪽 아래가 직각인 이등변 삼각형을 출력합니다.')
n = int(input('짧은 변 길이를 입력하세요.: '))

for i in range(n):              # 행 루프
    for _ in range(n - i - 1):  # 열 루프(공백을 출력)
        print(' ', end='')
    for _ in range(i + 1):
        print('*', end='')      # 열 루프(*을 출력)
    print()


#모든 행에서 공백과 *개수를 합하면 n
```
